Use NearestNeighbors to match data based on variables Z, given the file homework_1.2.csv.
Pick the best match in X = 0 corresponding to each X = 1. Use the Z values to perform the match: a good match with X = 1 is the item whose Z value is closest to the given sample's Z value with X = 0. I suggest using sklearn's NearestNeighbors to do this, but there are many ways to do it.  

```python
df_2 = pd.read_csv('/content/homework_1.2 (1).csv')
df_2.head()
```

```python
df_2.shape
```

```python
df_2.info()
```

```python
df_2.describe()
```

```python
df_2.isnull().sum()
```

```python
df_2.corr()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
X1_data = df_2[df_2['X'] == 1].reset_index(drop=True)
X0_data = df_2[df_2['X'] == 0].reset_index(drop=True)
```

```python
Z_columns = ['Z']

# Fit nearest neighbors on X=0 data
nn = NearestNeighbors(n_neighbors=1)
nn.fit(X0_data[Z_columns])

# Find closest match in X=0 for each row in X=1
distances, indices = nn.kneighbors(X1_data[Z_columns])
```

```python
farthest_distance = distances.max()
print("Farthest distance:", farthest_distance)
```

```python
matched_indices = indices.flatten()
matched_X0_Y = X0_data.iloc[matched_indices]['Y'].values
```

```python
avg_Y_X0_matched = matched_X0_Y.mean()
avg_Y_X1 = X1_data['Y'].mean()
effect = avg_Y_X1 - avg_Y_X0_matched
print("Effect:", effect)
```

```python
from collections import Counter

# Flatten all matched indices and count how many times each X=0 row was used
all_matched = np.concatenate(indices)
counts = Counter(all_matched)

# Count how many times each index appears MORE THAN ONCE
duplicates = sum([count - 1 for count in counts.values() if count > 1])
print("Duplicates:", duplicates)
```

```python
matched_group_means = []

for idx_list in indices:
    if len(idx_list) > 0:
        y_vals = X0_data.iloc[idx_list]['Y'].values
        matched_group_means.append(np.mean(y_vals))

# Mean of group means from X=0 matches
avg_matched_Y = np.mean(matched_group_means)

# Mean of Y for all X=1
avg_X1_Y = X1_data['Y'].mean()

effect = avg_X1_Y - avg_matched_Y
print("Effect:", effect)
```
